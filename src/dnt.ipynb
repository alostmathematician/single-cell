```python
import scanpy as sc
import scvi
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
DATA = Path("../data")
MODEL = Path("../model")
```

```python
vae_dnt = scvi.model.SCVI.load(MODEL / 'vae_dnt')
adata_dnt = vae_dnt.adata
```

```python
dnt_de = vae_dnt.differential_expression(groupby='cell_type')
```

```python
Tnfsf8 = dnt_de.loc['Tnfsf8']
```

```python
fig,ax = plt.subplots()
ax = sc.pl.violin(
    adata_dnt,keys='Tnfsf8',
    groupby='cell_type',
    ylabel='$Tnfsf8$ Expression Level Percentage(%)',
    ax=ax,
    show=False
    )

ax.set_xticklabels(['$aDNT$','$nDNT$'])
plt.show()
```
